```python
def transform(inputtext):
    
    res = (re.sub('.', lambda x: r'% 04X' % ord(x.group()), inputtext))
    package = ''
    numlist = []
    for i in range(1,len(res)):
        if(res[i] != ' '):
            package += res[i]
        else:
            numlist.append(int(package, 16))
            package = ''
    return numlist
```
